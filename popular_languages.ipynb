```python
df=df.drop(columns="Unnamed: 0")
```

```python
df['Average Annual Salary'] = df['Average Annual Salary'].str.replace('$','',regex=False).str.replace(',','')
df['Average Annual Salary'] = pd.to_numeric(df['Average Annual Salary'])
df1 = df.sort_values(by=['Average Annual Salary'],ascending=False)
df1.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))
import seaborn as sns
sns.set_theme(style='whitegrid')
sns.barplot(x="Average Annual Salary",y="Language", data=df1)
plt.title('Average Annual Salary by Language');
```
